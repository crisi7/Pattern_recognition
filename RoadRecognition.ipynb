```python
import cv2
import numpy as np
import csv
```

```python
# 全局变量
collecting_positive = False
collecting_negative = False
data = []

# 鼠标回调函数
def mouse_callback(event, x, y, flags, param):
    global collecting_positive, collecting_negative, data
    
    if event == cv2.EVENT_LBUTTONDOWN:  # 当鼠标左键被按下时
        # 获取点击位置的 BGR 值（OpenCV 默认使用 BGR 格式）
        b, g, r = image[y, x]
        print(f"Clicked at ({x}, {y}) - BGR: ({b}, {g}, {r})")

        # 根据当前采集模式将数据添加到列表中
        if collecting_positive:
            data.append((r, g, b, 'positive'))
        elif collecting_negative:
            data.append((r, g, b, 'negative'))

# 按键事件处理函数
def key_callback(key):
    global collecting_positive, collecting_negative, data
    
    if key == ord('y'):
        collecting_positive = True
        collecting_negative = False
        print("Collecting positive samples...")
    elif key == ord('n'):
        collecting_positive = False
        collecting_negative = True
        print("Collecting negative samples...")
    elif key == ord('e'):
        collecting_positive = False
        collecting_negative = False
        print("Ending collection. Collected data:")
        for d in data:
            print(d)
        return False  # 结束循环
    return True  # 继续循环
```

```python
# 图片路径
image_path = './data/0618.png'

# 读取图片
image = cv2.imread(image_path)

# 检查图片是否成功读取
if image is None:
    print("Error: Could not read the image. Check the file path.")
else:
    # 创建一个窗口
    window_name = "Image"
    cv2.namedWindow(window_name)
    
    # 设置鼠标回调函数
    cv2.setMouseCallback(window_name, mouse_callback)

    # 显示图片并处理按键事件
    while True:
        cv2.imshow(window_name, image)
        key = cv2.waitKey(1) & 0xFF
        
        # 处理按键事件
        if not key_callback(key):
            break

    # 关闭所有 OpenCV 窗口
    cv2.destroyAllWindows()

    # 打印收集到的数据
    print("\nCollected Pixel Data:")
    for d in data:
        print(d)

    # 保存数据到 CSV 文件
    with open('sample_data.csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['R', 'G', 'B', 'Label'])
        for d in data:
            writer.writerow(d)

    print("Data saved to sample_data.csv")
```
